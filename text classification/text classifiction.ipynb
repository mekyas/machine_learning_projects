## Split labelled data into training and validation data set. Then build your model from training data and evaluate your prediction on validation data using suitable metric.

The number of observation per label is different.
Class 3 has aproximatly 45% of data
So we must make sure to split the data in a stratified manner and to use f1_score as a metric instead of accuracy

```python
Xtrain, Xval, y_train, y_val = train_test_split(train.token.values, 
                                                train.label, 
                                                train_size=0.9,
                                                stratify = train.label,
                                                random_state=123)
```

```python
# to make sure the classes are label from zero
# if not, some machine learning libraries will suppose that there is 5 classes
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_val = encoder.fit_transform(y_val)
```

```python
# transform the data to a matrix of count of each word
count_vect = CountVectorizer(analyzer='word', max_features= 10000)
count_vect.fit(train.token)
Xtrain_count = count_vect.transform(Xtrain)
Xval_count = count_vect.transform(Xval)
# reduce dimension for faster training 
svd = TruncatedSVD(50, random_state=42) 
Xtrain_count = svd.fit_transform(Xtrain_count) 
Xval_count = svd.transform(Xval_count)

# N-grams level of TF-IDF
tfidf_ngram = TfidfVectorizer(analyzer='word', 
                              max_features= 10000, 
                              ngram_range=(2, 3))
tfidf_ngram.fit(train.token)
Xtrain_tfidf_ngram = tfidf_ngram.transform(Xtrain)
Xval_tfidf_ngram = tfidf_ngram.transform(Xval)

svd = TruncatedSVD(50, random_state=42) 
Xtrain_tfidf_ngram = svd.fit_transform(Xtrain_tfidf_ngram) 
Xval_tfidf_ngram = svd.transform(Xval_tfidf_ngram)

# words level of TF-IDF
vectorizer_params = {
    'ngram_range': (1, 1),
    'max_df': 0.5,
    'max_features':10000,
}
tfidf_word = TfidfVectorizer(**vectorizer_params)
tfidf_word.fit(train.token)
Xtrain_tfidf_word = tfidf_word.transform(Xtrain)
Xval_tfidf_word = tfidf_word.transform(Xval)

svd = TruncatedSVD(50, random_state=42) 
Xtrain_tfidf_word = svd.fit_transform(Xtrain_tfidf_word) 
Xval_tfidf_word = svd.transform(Xval_tfidf_word)
```

```python
def train_model(xtrain, ytrain, xtest, ytest, model=None, 
                params = None, model_type= 'sklearn'):
    """Help function to train different types of model
    and print out the f1 score
    Argument:
        xtrain: (numpy array) training data
        ytrain: (numpy array) training labels
        xtest: (numpy array) validation data
        ytest: (numpy array) validation labels
        model: sklean model object 
        params: (dictionary) parameters of the model
        model_type : (string) type of the model to train
                    possible values are 'sklearn, xgb, cat, lgb
    Return
        object of the trained model
    """
    test_pred = np.zeros_like(xtest)
    if model_type == 'sklearn':
        model.fit(xtrain, ytrain)
        pred = model.predict(xtrain)
        test_pred = model.predict(xtest)
    
    if model_type == 'xgb':
        eval_set=[(xtrain, ytrain), (xtest, ytest)]
        model = xgb.XGBClassifier(**params, silent=False, )
        model.fit(xtrain, ytrain, eval_set=eval_set, early_stopping_rounds= 10)
        pred = model.predict(xtrain, ntree_limit=model.best_ntree_limit)
        test_pred = model.predict(xtest, ntree_limit=model.best_ntree_limit)

    if model_type == 'cat':
        model = cat.CatBoostClassifier(**params)
        model.fit(xtrain, 
                  ytrain, 
                  verbose=False,
                  use_best_model = True,
                 eval_set=(xtest, ytest))
        pred = model.predict(xtrain)
        test_pred = model.predict(xtest).reshape(-1,)

    if model_type == 'lgb':
        dtrain = lgb.Dataset(xtrain, label=ytrain)
        dtest = lgb.Dataset(xtest, label= ytest)
        model = lgb.train(params, 
                          dtrain, 
                          valid_sets=[dtrain, dtest],
                          early_stopping_rounds=100)
        pred = np.argmax(model.predict(xtrain, num_iteration=model.best_iteration), axis=1)
        test_pred = np.argmax(model.predict(xtest, num_iteration=model.best_iteration), axis = 1)
    
    # Since the taget variable is imbalanced, we will use F1_score as metrics
    train_score = f1_score(ytrain, pred, average='weighted')
    print("The f1_score for training set is {:.4f}".format(train_score))
    if ytest is not None:
        print("The f1_score for validation set is {:.4f}".format(f1_score(ytest, test_pred, average='weighted')))
    print()
    return model
```

```python
def plot_history(history):
    """function to plot the loss and the accuracy 
    of the training set and test set of a neural network
    
    Arguments: the history object of the fitted model"""
    
    fig, ax = plt.subplots(2, 1, figsize = [10, 8])
    ax[0].plot(history.epoch, history.history["loss"], c = 'b', label='training loss')
    ax[0].plot(history.epoch, history.history["val_loss"], '.b', label='validation loss')
    ax[0].set_ylabel("Loss")
    ax[0].set_title("Evolution of loss during training")
    ax[0].legend()
    
    ax[1].plot(history.epoch, history.history["acc"], c = 'b', label='training accuracy')
    ax[1].plot(history.epoch, history.history["val_acc"], '.b', label='validation accuracy')
    ax[1].set_xlabel("number of epochs")
    ax[1].set_ylabel("Accuracy")
    ax[1].set_title("Evolution of accuracy during training")
    ax[1].legend()
```
